To find items that don't have a corresponding restaurant in restaurant.csv, you would perform a left join (or merge) using the item.csv as the left dataframe and restaurant.csv as the right dataframe on the respective keys. Then, you'd filter the resulting dataframe where the restaurant details are missing (NaN).

```python
# Perform a left merge
merged_df = pd.merge(item_df, rest_df, left_on='restaurant_id', right_on='id', how='left')
```

1. Functions in Pandas for Data Analysis:
a) Descriptive Statistics:
describe(): Provides a summary of the main statistics (like count, mean, std, min, and max).
mean(), median(), mode(): Calculate the mean, median, and mode.
sum(): Sum of values.
min(), max(): Minimum and maximum values.
var(): Variance of values.
std(): Standard deviation of values.
corr(): Correlation matrix of values.
cov(): Covariance matrix of values.

b) Data Filtering and Sorting:
head(), tail(): Display the first or last few rows of the dataframe.
loc[], iloc[]: Slice or filter rows and columns by label or index.
query(): Query the data based on conditional expressions.
sort_values(): Sort by the values along either axis.
sort_index(): Sort dataframe by indices.

c) Data Transformation:
groupby(): Group dataframe using a particular column, helpful for aggregate operations.
pivot_table(): Create a spreadsheet-style pivot table as a dataframe.
melt(): Unpivot a dataframe from wide to long format.
crosstab(): Compute a simple cross tabulation of two (or more) factors.

d) Missing Data:
isnull(), notnull(): Detect missing values.
fillna(): Fill missing values.
dropna(): Drop missing values.

e) Others:
unique(): Get unique values.
nunique(): Count of unique values.
value_counts(): Return a series containing counts of unique values.
apply(): Apply a function along the axis of the dataframe.
map(): Map values of series using input correspondence.

2. Join Commands in Pandas:
merge(): Similar to SQL joins, it allows joining on columns or indices.
how parameter options:
inner: Use intersection of keys from both frames, similar to a SQL inner join; preserve the order of the left keys.
outer: Use union of keys from both frames, similar to a SQL full outer join; sort keys lexicographically.
left: Use keys from the left frame only, similar to a SQL left outer join; preserve key order.
right: Use keys from the right frame only, similar to a SQL right outer join; preserve key order.
join(): Join columns of another dataframe. It's a more simplified function for joining on indices.
concat(): Concatenate pandas objects along a particular axis.

The df.query() method in pandas allows you to query the data in a DataFrame using a string expression. It's a concise way to filter data based on conditions.

Let's look at how to use the query() method for the restaurant_df DataFrame based on various example scenarios:



```python
#Filtering restaurants that were opened after April 2015:
opened_after_april_2015 = rest_df.query("open_date > '2015-04-30'")
opened_after_april_2015.head()
```

```python
#Filtering restaurants with a rating greater than 4:
high_rated_restaurants = rest_df.query("rating > 4")
high_rated_restaurants.head()
high_rated_restaurants.shape[0]
```

```python
#Filtering restaurants that offer delivery and are verified:
delivery_and_verified = rest_df.query("delivers == 1 and verified == 1")
delivery_and_verified.head()
```

```python
#Filtering restaurants based in NSW state using the address column:
nsw_restaurants = rest_df.query("', NSW,' in address")
nsw_restaurants.head()
```

```python
#Filtering restaurants opened in the year 2015
opened_in_2015 = rest_df.query("open_date >= '2015-01-01' & open_date <= '2015-12-31'")
opened_in_2015.head()
```
